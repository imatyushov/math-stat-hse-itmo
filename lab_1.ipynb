## 1) Сгенерировать выборку из равномерного распределения на отрезке [0, 1]

```python
import pandas as pd;
import numpy as np;
from matplotlib import pyplot as plt;
import random;
import math;
from scipy.stats import uniform;
```

```python
sample_size = 100;
sample_selection = np.random.uniform(0, 1, sample_size);
print(sample_selection);
first_moment = np.mean(sample_selection)
print('fm', first_moment)
```

```python
sorted_sample = np.sort(sample_selection) #sort
print(sorted_sample)
```

## 2) Написать процедуру построения выборочной функции распределения по выборке (график)

```python
def create_selection_function(sample_selection):
  sorted_sample = np.sort(sample_selection) #sort
  sample_size = len(sample_selection) #объем выборки
  y = np.arange(1, sample_size + 1) / sample_size
  print(f'Sample size: { sample_size }')
  return sorted_sample, y


x, y = create_selection_function(sample_selection)

print(f'x: {x} y: {y}')

plt.step(x, y, where='post')
plt.plot(x, x, color='red', linestyle='--')  # adding y=x line to the plot
plt.xlabel('Value x')
plt.ylabel('Выборочная ф-ция распределения Fr(x)')
plt.title('Selective distribution function')
plt.grid = True
plt.show()
```
